Preprocessing adult.data
=====================

This script helps to preprocess `adult.data` that can be downloaded from the original [UCI Machine Learning repository](https://archive.ics.uci.edu/ml/datasets/adult).

```python
from pathlib import Path

import pandas as pd
```

**TODO**: please configure `DATA_PATH` and `RAW_ADULT_DATASET_PATH`.

```python
DATA_PATH = Path('/home/nap/Workspace/Privacy-DARC/rmckenna-adult/data')
RAW_ADULT_DATASET_PATH = DATA_PATH / Path('adult.data')
OUTPUT_ADULT_DATASET_PATH = DATA_PATH / Path('adult.csv')

COLUMN_NAMES = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
                'salary']

df = pd.read_csv(RAW_ADULT_DATASET_PATH, names=COLUMN_NAMES)
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df.sample(10)
```

```python
df.to_csv(OUTPUT_ADULT_DATASET_PATH, index=False)
```
